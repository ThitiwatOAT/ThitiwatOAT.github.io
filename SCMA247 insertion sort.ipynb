```python
def InsertionSort(array):

    for x in range (1, len(array)):
        for i in range(x, 0, -1):
            if array[i] < array[i - 1]:
                t = array[i]
                array[i] = array[i - 1]
                array[i - 1] = t
            else:
                break
            i = i - 1
    return array
```

```python
def InsertionSortWExp(arr):
    for x in range (1, len(arr)):
        print('x =',x)
        # backward loop from end to start
        for i in range(x, 0, -1):
            print('i =',i)

            # linearly compare backward every elt
            # arr[i]=right
            # arr[i-1]=left
            print('   right =',arr[i],'< left =',arr[i-1],'if true, then swap')
            if arr[i] < arr[i - 1]:
                print('   ',"it's ✅, means bigger is on the left side which is wrong, so swap")

                # swap process
                print('       before',arr)
                temp = arr[i] # store right = smallerNum
                arr[i] = arr[i - 1] # right = left = biggerNum
                arr[i - 1] = temp # left = temp = smallerNum
                print('       after ',arr)
            else:
                print('   ',"it's ❌, already correct, then break")
                break
            i = i - 1
            print('- - - - - - - - - - - - - - - -')
        print('____________________________________________________________')
```
